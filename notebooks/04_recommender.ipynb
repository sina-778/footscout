```
import sys; sys.path.insert(0,"..")
import pandas as pd
from src.embeddings import load_embeddings
from src.recommender import FootScoutRecommender, get_radar_data, make_radar_figure
df = pd.read_csv("../data/processed/players_merged.csv", low_memory=False)
embs = load_embeddings(method="umap", alpha=0.6)
rec = FootScoutRecommender(df=df, embeddings=embs["hybrid"])
PLAYER = df["player"].iloc[0]  # Change to any player name
print(f"Ready — query: {PLAYER}")
```

## Mode 1 — Standard Similarity

```
results = rec.find_similar(PLAYER, k=5)
results[["rank","player","similarity","position","squad","market_value_eur"]]
```

## Mode 2 — Budget-Aware Replacement

```
budget = rec.find_budget_replacement(PLAYER, budget=30_000_000, k=5)
budget[["rank","player","similarity","market_value_eur"]]
```

## Mode 3 — Hidden Gem Finder

```
gems = rec.find_hidden_gems(position="FW", max_value=15_000_000, k=5)
gems[["rank","player","similarity","market_value_eur"]]
```
